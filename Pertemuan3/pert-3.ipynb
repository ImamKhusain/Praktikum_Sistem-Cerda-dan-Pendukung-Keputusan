## GRAFIK DAN PYTHON SCRIPTS 

```python
import matplotlib.pyplot as plt  
x = [1, 2, 3]  
y = [3, 2, 6]

plt.title(
    "Grafik Penjualan",
    fontsize = 18,
    color = "blue",
    fontweight = "bold",
    loc="left",
    pad= 18,
)
plt.ylabel("Profit")
plt.xlabel(
    "Rentang Waktu",
    fontsize = 12,
    color = "black",
    fontweight = "bold",
    fontstyle = "italic",
    fontfamily = "calibri",
    loc="center",
    labelpad= 20,
)
plt.plot(x, y) 
plt.show() 
```

```python
x = [1, 2, 3]  
y = [3, 2, 6]
plt.plot(
    x,
    y,
    "b-.*",
    markersize= 14,
)

plt.show()
```

### plt grid

```python
plt.plot(x, y)
plt.grid(
    True, 
    alpha = 0.3, 
    color = "red", 
    linestyle="--", 
    linewidth = 1
)
plt.show()
```

### LEGEND

```python
plt.plot(x,y, label="penjualan")
plt.plot(x,[3,4,5], label="pembelian")
plt.plot(x,[7,3,5], label="profit")
plt.grid(True, alpha = 0.5, color='red', linestyle='--', linewidth=1)
plt.legend(title="ini lgend", facecolor="yellow", edgecolor='red', shadow=True, ncol=2, loc="upper center", fontsize="small")
plt.show()
```

```python

x = [1, 2, 3]
y = [4, 5, 6]
b = [7, 8, 9]
a = [3, 5, 8]
plt.plot(x, y, color='green', linestyle='--', marker='o') # plt.plot(x, y, 'r--o') shortcutnya/pendeknya
plt.plot(a, b, 'r--o')
plt.title('Grafik Sederhana', fontsize=30, pad=20, loc='center', fontweight='bold', color="blue", fontstyle='italic')
plt.xlabel('Sumbu X', labelpad=20, loc='center')
plt.ylabel('Sumbu Y',labelpad=20, loc='center')
plt.show()
```
